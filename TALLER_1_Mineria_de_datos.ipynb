**<h1> FRECUENCIA ABSOLUTA </h1>**

<h2>La frecuencia absoluta indica el numero de veces que un valor aparece en un conjunto de datos. Para calcularla utilizamos la función 'pd.value_counts()'.</h2>

<h2>En este caso usamos las tres columnas cualitativas de nuestro dataframe.</h2>

```python
f_absoluta=pd.value_counts(df['MUNICIPIO'])
f_absoluta
```

```python
f_absoluta=pd.value_counts(df['DEPARTAMENTO'])
f_absoluta
```

```python
f_absoluta=pd.value_counts(df['PRODUCTO'])
f_absoluta
```

**<h1> FRECUENCIA RELATIVA</h1>**

<h2>Para calcular la frecuencia relativaa de un conjunto de datos se divide la frecuencia absoluta de caa valor por el número total de valores en el conjunto de datos. Usando la función 'value_counts' la cual sirve ara contar la frecuencia absoluta realizamos el calculo.</h2>

<h2>El resultado que recibimos son valores decimales, el cual representan un porcentaje. Es decir, en el primer caso la variable 'MUNICIPIO', el dato 'BOGOTA, DC' el valor es de 6...., este se representa como un 60%. Así con el resto de valores, y la suma de todas las frecuencias relativas simpre será 100% .</h2>

```python
f_relativa = 100* df['MUNICIPIO'].value_counts() / len(df['MUNICIPIO'])
f_relativa
```

```python
f_relativa = 100* df['DEPARTAMENTO'].value_counts() / len(df['DEPARTAMENTO'])
f_relativa
```

```python
f_relativa = 100* df['PRODUCTO'].value_counts() / len(df['PRODUCTO'])
f_relativa
```

**<h1> GRAFICO DE BARRAS </h1>**

<h2>La función value_counts() devuelve una serie de pandas con los valores únicos y su frecuencia. Luego, se utiliza la función plot(kind='bar') para crear el gráfico de barras. El parámetro kind='bar' indica que se desea un gráfico de barras..</h2>

```python

grafico_1 = df['PRODUCTO'].value_counts().plot(kind = 'bar', title = 'PRODUCTO')
grafico_1
```

<h2>En este caso, se utiliza el método len( ) para contar el número total de observaciones en la serie de datos. Luego, se calculan los porcentajes dividiendo cada valor de la frecuencia por el total y multiplicando por 100. Finalmente, se utiliza la función plot(kind='bar') para crear el gráfico de barras con los porcentajes.
.</h2>

```python
grafico_2 = plot = (100 * df['PRODUCTO'].value_counts() / len(df['PRODUCTO'])).plot(kind = 'bar', title = '% PRODUCTO')
grafico_2
```

**<h1> GRAFICO DE TORTA </h1>**


<h2>Contar la frecuencia de cada valor utilizando value_counts().</h2>

<h2>En este caso, se utiliza la función plot(kind='pie') para crear el gráfico de torta con los porcentajes de cada valor. El parámetro autopct se utiliza para mostrar los porcentajes en cada sector </h2>

```python
grafico_3 = df['PRODUCTO'].value_counts().plot(kind='pie', autopct = '%2.f', figsize = (10, 10), title = 'PRODUCTO')
grafico_3
```
